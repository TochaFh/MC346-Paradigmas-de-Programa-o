```racket
;(require mzlib/defmacro)
;(require (for-syntax racket/base))

; fazendo "else" ser capturado como uma variável de valor #t
(define-macro (acond* . clauses)
  (unless (null? clauses)
    (let ([cl1   (car clauses)]
          [gcond (gensym "cond")])
      `(let ([,gcond ,(car cl1)]
             [else #t])
         (if ,gcond
             (let ([it ,gcond])
               ,@(cdr cl1))
             (acond ,@(cdr clauses)))))))


; conferindo a palavra else de forma mais segura
(define-macro (acond . clauses)
  (unless (null? clauses)
    (let ([cl1   (car clauses)]
          [gcond (gensym "cond")])
      (if (eq? (car cl1) 'else)
          `(begin ,@(cdr cl1))
          `(let ([,gcond ,(car cl1)])
              (if ,gcond
                  (let ([it ,gcond])
                    ,@(cdr cl1))
                  (acond ,@(cdr clauses))))))))
  

(acond
     ((> 2 1) (display "hehe"))
     (else (display "else"))
)
```

```racket
(require mzlib/defmacro)
(require (for-syntax racket/base))

(define-macro (for-1 vvs . body)
  (let ([var    (car vvs)]
        [start  (cadr vvs)]
        [stop   (caddr vvs)]
        [glimit (gensym "limit")])
    (printf "GLIMIT: ~A\n" glimit)
    `(do ((,var ,start (+ 1 ,var))
          (,glimit ,stop))
         ((> ,var ,glimit))
       ,@body)))

(for-1 (limit 1 5)
  (display limit))
```
