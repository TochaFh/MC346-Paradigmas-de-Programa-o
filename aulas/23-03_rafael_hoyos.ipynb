## Exercício 1

Defina uma função que receba como parâmetro um número e retorne o valor absoluto do número. Escreva um exemplo que use esta função.

````scheme
(define (absoluto num)
  (if (< num 0)
      (- 0 num)
      num
  )
)
````

## Exercício 2

Escreva uma função chamada "regressiva" que receba um parâmetro inteiro positivo "n" e retorne uma lista com a contagem regressiva de "n" até 1. Para simplificar, não é necessário verificar se "n" é um inteiro positivo. Escreva um exemplo que gere a lista regressiva de 5.

Considere usar recursão como forma de repetição.

````scheme
(define (regressiva num)
  (if (> num 0)
      (cons num (regressiva (- num 1)))
      '()
      
  )
)
(regressiva 5)
````

# `car` e `cdr`

## Exercício 3

Construa uma solução para retornar o segundo elemento de uma lista.

````scheme
(define (second l)
  (car (cdr l))
)
(second '(10 20 30 40))
````

## Exercício 4

Defina uma função `replace-first` que receba um elemento P e uma lista L, e substitua o primeiro elemento de L por P. Elabore um exemplo com `replace-first`.

````scheme
(define (replace-first P L)
  (cons P (cdr L))
)
(replace-first 67 '(10 20 30 40 50))
````

## Exercício 5

Dada uma estrutura do tipo:
```scheme
(define pessoa
  '((Asdrúbal 25)
    (Doriana 30)
    (Quincas 28)))
```
Escreva uma função que receba a estrutura e gere uma lista apenas com o segundo elemento (lista de idades).

````scheme
(define pessoa
  '((Asdrúbal 25)
    (Doriana 30)
    (Quincas 28)))

(define (col_2 l)
  (if (null? l)
    '()
    (cons (cadr (car l)) (col_2 (cdr l)))
  )
)

(col_2 pessoa)
````

## Exercício 6

Quebre a função anterior em duas funções:
* função `second`: recebe uma lista com dois elementos e retorna o segundo elemento
* função `process`: recebe dois parâmetros -- uma lista de listas e uma função; ela aplica a função na lista de listas, por exemplo, se a função for `second` ela extrai o segundo elemento de todas as sublistas.

````scheme
(define pessoa
  '((Asdrúbal 25)
    (Doriana 30)
    (Quincas 28)))

(define (second l)
  (car (cdr l))
)

(define (col_2 process l)
  (if (null? l)
    '()
    (cons (process (car l)) (col_2 process (cdr l)))
  )
)

(col_2 second pessoa)
````

````scheme
(col_2 cadr pessoa)
````

````scheme
(col_2 car pessoa)
````
