## Exercício A

Dada uma estrutura do tipo:

```scheme
(define friends-list
  '((Asdrúbal 25)
    (Doriana 30)
    (Quincas 28)))
```

Escreva uma função que receba a estrutura e gere uma lista apenas com o segundo elemento (lista de idades).

```scheme
(define (ages people-list)
  (if (null? people-list)
      '()
      (cons (cadr (car people-list))
            (ages (cdr people-list)))))

(ages friends-list)
```

# Função de Ordem Superior

## Exercício B

Quebre a função anterior em duas funções:
* função `second`: recebe uma lista com dois elementos e retorna o segundo elemento
* função `process`: recebe dois parâmetros -- 

```scheme
(define (second item)
  (cadr item))

(define (process items-list func)
  (if (null? items-list)
      '()
      (cons (func (car items-list))
            (process (cdr items-list) func))))

(process friends-list second)
```

# `map` - função de ordem superior

```scheme
(map second friends-list)
```

## Exercício 2

Dada a seguinte lista:

```scheme
(define employee-list
  '((1012 Melissa 41)
    (1717 Asdrúbal 25)
    (1921 Doriana 30)
    (2549 Quincas 28)
    (3456 Bonerges 75)))
```


Em cada sublista, o primeiro elemento é um código, o segundo, o nome e o terceiro, a idade.

A seguir, foi construído um processo em três etapas. Você deve descrever brevemente abaixo de cada etapa o que ela faz.

```scheme
(define (extract n)
  (lambda (lst)
    (if (= n 1)
        (car lst)
        ((extract (- n 1)) (cdr lst)))))
```

> descreva aqui
> A função extract (recebe n como parâmetro) vai retornar uma função (que recebe uma lista) que retorna o enésimo elemento de uma lista.
> O lambda é usado para fazer a "definição" dessa função dinâmica que vai depender de n.

```scheme
(define third (extract 3))
```

> descreva aqui
> Define 'third' como uma funçao que retorna o terceiro elemento de uma lista (a partir do extract).

```scheme
(map third employee-list)
```

> descreva aqui
> Usa o map com third para obter uma lista com as idades dos employees.

## Exercício 3

Escreva uma única função `age` que receba uma lista com estrutura equivalente à `employee-list` como parâmetro, mais um código, e retorne a idade da pessoa que tem aquele código.

Alguns requisitos:
1) Para simplificar, a função sempre considera que existe o elemento na lista. Não é necessária a verificação.
2) Você deve usar nesta solução duas funções de alta ordem apresentadas (uma que gera uma função e outra que recebe uma função), bem como a função lambda.

```scheme
; ----- employee-list e extract ----

(define employee-list
  '((1012 Melissa 41)
    (1717 Asdrúbal 25)
    (1921 Doriana 30)
    (2549 Quincas 28)
    (3456 Bonerges 75)))

(define (extract n)
  (lambda (lst)
    (if (= n 1)
        (car lst)
        ((extract (- n 1)) (cdr lst)))))


; ---- feito na "gambiarra" ----

(define (get_element_by_key lst key_getter key)
  (if (= (key_getter (car lst)) key)
      (car lst)
      (get_element_by_key (cdr lst) key_getter key)
  )
)

(define (age lst code)
  ((extract 3) (get_element_by_key lst (extract 1) code))
)

(age employee-list 2549)
```

```scheme
; ---- usando closure (usando o find) ----

(define (compara_code code)
  (lambda (lst) (= code (car lst)))
)

(define (age lst code)
  ((extract 3) (find (compara_code code) lst))
)

(age employee-list 2549)
```

```scheme
(define (make-counter step)
  (let ((n 0))
    (lambda ()
      (let ((current n))
        (set! n (+ n step))
        current))))
```
