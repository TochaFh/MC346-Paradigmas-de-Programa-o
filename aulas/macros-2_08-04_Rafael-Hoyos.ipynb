#### Questão 1
Macro "concatenação" de listas.

````scheme
(define x '(1 2 3))
(define y '(3 2 1))

(define-macro (concat a b)
  `'(,@a ,@b))

(display (concat (1 2 3) (3 2 1)))


(define-macro (concat-s a b)
  `(append ,a ,b))

(display (concat-s x y))


(define-macro (concat-d a b)
  (list append a b))

(display (concat-d x y))
````

#### Questão 2
for-1

````scheme
(define-macro (for-1 vvs . body)
  (let ([i   (car vvs)]
        [i0  (cadr vvs)]
        [il (cadr (cdr vvs))])
    `(do ((,i ,i0 (1+ ,i)))
         ((> ,i ,il))
       ,@body)))

(for-1 (x 1 5)
  (display x))

(for-1 (x 50 57)
  (display "\n")
  (display x))
````

#### Questão 3
O que o código a seguir faz?

```scheme
(for-1 (limit 1 5)
  (display limit))
``` 

Printa os números de 1 a 5 em sequência "12345".
